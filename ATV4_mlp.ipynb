```python
import numpy as np
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
## Exercício 1 - Problema do XOR
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])

# 1- Importar o modelo MLP do sklearn
from sklearn.neural_network import MLPClassifier

# 2- Instanciar o modelo escolhendo uma topologia para a rede, função de ativação e número de épocas de execução até 100% de acerto
model_xor = MLPClassifier(hidden_layer_sizes=(4, 3), activation="tanh", max_iter=4000)

# 3- Treinar o modelo com os dados de treinamento
model_xor.fit(x, y)

# 4- Fazer o predict com os dados de treinamento
model_xor.predict(x)

# 5- Comparar o resultado do predict com o vetor y
from sklearn import metrics
result_xor = model_xor.predict(x)
acc_xor = metrics.accuracy_score(result_xor, y)
show_xor = round(acc_xor * 100)
print("{}%".format(show_xor))

print(list(result_xor))
print(list(y))
```

```python
## Exercício 2 - Dataset Parkinsons

# 1- Carregar a base de dados Parkinsons
url_2 = "https://raw.githubusercontent.com/tmoura/machinelearning/master/parkinsons.data"
df_2 = pd.read_csv(url_2, header=None)
columns_2 = len(df_2.columns)
df_2
```

```python
# 2- Normalizar todas as colunas (normalizar é deixar todos os valores das colunas entre 0 e 1)
from sklearn.preprocessing import MinMaxScaler

for c in range(1, 23):
  df_2[c] = MinMaxScaler().fit_transform(np.array(df_2[c]).reshape(-1, 1))
df_2
```

```python
# 3- Separar o dataset em X (matriz de features) e y (coluna target)
y_2 = df_2[0] #target
X_2 = df_2.loc[:, 1:columns_2-1] #antes da vírgula todas as linhas, depois da vígula todas as colunas a partir da de índice 1 até a última

# 4- Gerar as bases de treinamento e teste
X2_train, X2_test, y2_train, y2_test = train_test_split(X_2, y_2, test_size=0.2, random_state=None, stratify=y_2)

# 5- Importar o modelo MLP do sklearn
from sklearn.neural_network import MLPClassifier

# 6- Instanciar o modelo escolhendo uma topologia para a rede, função de ativação e número de épocas de execução até que obtenha uma taxa de acerto estável
model_2 = MLPClassifier(hidden_layer_sizes=(4, 3), activation="tanh", max_iter=4000)

# 7- Treinar o modelo com os dados de treinamento
model_2.fit(X2_train, y2_train) # treinamento do modelo com os dados x2 e y2 _train

# 8- Fazer o predict com os dados de teste
model_2.predict(X2_test) #predição de resultado

# 9- Imprimir o percentual de acerto da base de teste
result_2 = model_2.predict(X2_test)
acc_2 = metrics.accuracy_score(result_2, y2_test)
show_2 = round(acc_2 * 100)
print("{}%".format(show_2))

print(list(result_2))
print(list(y2_test))
```

```python
import seaborn as sns
# 1- Carregar a base de dados Penguins da API do Seaborn
df_3 = sns.load_dataset("penguins")

# 2- A base precisará ser tratada: Existem valores nulos e dados categóricos
df_3.dropna(axis=0, inplace=True) #retira as linhas com valores nulos

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import LabelEncoder

#transformação dos dados categóricos em numéricos com o OneHotEncoder
column_tranformer = make_column_transformer((OneHotEncoder(), ['island', 'sex']), remainder='passthrough')
df_3 = column_tranformer.fit_transform(df_3)
df_3 = pd.DataFrame(data=df_3)
#transformação do target com Label Encoder
label_encoder = LabelEncoder()
df_3[5] = label_encoder.fit_transform(df_3[5])

# 3- Normalizar todas as colunas (normalizar é deixar todos os valores das colunas entre 0 e 1) --> Redes Neurais são sensíveis as diferenças escalares de valores das features
from sklearn.preprocessing import MinMaxScaler
for c in range(6, 10):
  df_3[c] = MinMaxScaler().fit_transform(np.array(df_3[c]).reshape(-1, 1))
df_3

# 4- Separar o dataset em X (matriz de features) e y (coluna target)

y_3 = df_3[5]
X_3 = df_3.drop(5, axis=1)
```

```python
# 5- Gerar as bases de treinamento e teste
X3_train, X3_test, y3_train, y3_test = train_test_split(X_3, y_3, test_size=0.2, random_state=None, stratify=y_3)

# 6- Importar o modelo MLP do sklearn
from sklearn.neural_network import MLPClassifier
# 7- Instanciar o modelo escolhendo uma topologia para a rede, função de ativação e número de épocas de execução até que obtenha uma taxa de acerto estável
model_3 = MLPClassifier(hidden_layer_sizes=(10, 5), activation='tanh', max_iter=1500)

# 8- Treinar o modelo com os dados de treinamento
model_3.fit(X3_train, y3_train)
# 9- Fazer o predict com os dados de teste
model_3.predict(X3_test)

# 10- Imprimir o percentual de acerto da base de teste
result_3 = model_3.predict(X3_test)
acc_3 = metrics.accuracy_score(result_3, y3_test)
show_3 = round(acc_3 * 100)
print("{}%".format(show_3))

print(list(result_3))
print(list(y3_test))
```
