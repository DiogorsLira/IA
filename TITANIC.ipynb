Árvore de decisão e KNN TITANIC

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn import metrics
import pandas as pd
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
```

```python
# carregando dataset TITANIC
df = sns.load_dataset('titanic')
df
```

```python
#quantidadde de NaN em deck
qtd_nan = df['deck'].isna().sum()
qtd_nan
```

```python
#removendo as linhas com NaN e classe class por ser igual a pclass
df.drop('class', axis=1, inplace=True) #para retirar a class, inplace=true evita o df = df...
df.drop('deck', axis=1, inplace=True) #retirar deck devido a enorme quantidade de linhas sem valores
df.drop('alive', axis=1, inplace=True) #retirar alive pq é igual a survived, sendo alive categórico e survived numérico

df.dropna(axis=0, inplace=True) #retira todos os NaN nas linhas

df
```

```python
#transformando os dados categóricos em numéricos com o OneHotEncoder
column_transformer = make_column_transformer((OneHotEncoder(), ['sex', 'embarked', 'who', 'adult_male', 'embark_town', 'alone']), remainder='passthrough')

df = column_transformer.fit_transform(df)

#columns_names = column_transformer.get_feature_names_out() caso queira adicionar os nomes nas colunas (adc: columns = columns_names em df abaixo)

df = pd.DataFrame(data = df)

df
```

```python
#definindo o resultado e dados para o treinamento
columns = len(df.columns)
y = df[15] #classe survived
X = df.drop(15, axis=1) #todas menos a survived
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=None, stratify=y) # 20% para teste

#treinamendo da Árvore de Decisão
model_entropy = tree.DecisionTreeClassifier(criterion='entropy')
model_entropy = model_entropy.fit(X_train, y_train)

model_gini = tree.DecisionTreeClassifier(criterion='gini')
model_gini = model_gini.fit(X_train, y_train)

#treinamento para knn
k = 3 #vizinhos próximos

model_euclidean = KNeighborsClassifier(n_neighbors=k, metric='euclidean', algorithm='brute')
model_euclidean = model_euclidean.fit(X_train, y_train)

model_manhattan = KNeighborsClassifier(n_neighbors=k, metric='manhattan', algorithm='brute')
model_manhattan = model_manhattan.fit(X_train, y_train)
```

```python
#predição e resultados

#para entropia
result_entropy = model_entropy.predict(X_test)
acc_entropy = metrics.accuracy_score(result_entropy, y_test)
show_entropy = round(acc_entropy * 100)

#para o gini
result_gini = model_gini.predict(X_test)
acc_gini = metrics.accuracy_score(result_gini, y_test)
show_gini = round(acc_gini * 100)

#para distancia euclidiana
result_euclidean = model_euclidean.predict(X_test)
acc_euclidean = metrics.accuracy_score(result_euclidean, y_test)
show_euclidean = round(acc_euclidean * 100)

#para distancia manhattan
result_manhattan = model_manhattan.predict(X_test)
acc_manhattan = metrics.accuracy_score(result_manhattan, y_test)
show_manhattan = round(acc_manhattan * 100)

print("Árvore com cálculo de entropia")
print(list(result_entropy))
print(list(y_test))
print("{}%".format(show_entropy))

print("Árvore com cálculo de gini")
print(list(result_gini))
print(list(y_test))
print("{}%".format(show_gini))

print("KNN-Euclidean")
print(list(result_euclidean))
print(list(y_test))
print("{}%".format(show_euclidean))

print("KNN-Manhattan")
print(list(result_manhattan))
print(list(y_test))
print("{}%".format(show_manhattan))
```
