```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pandas as pd
from
```

```python
#primeiro dataset / raisin
url = "/content/sample_data/Raisin_Dataset.csv"

# Carregar base de dados
# DataFrame
df = pd.read_csv(url)

columns = len(df.columns)
#separação dos dados em target(y) e os outros x
y = df["Class"]
X = df.drop("Class", axis=1)

#separando os dados em teste e treinamento
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None, stratify=y) # 80% treino e 20% teste

# Treinamendo da Árvore de Decisão
model = tree.DecisionTreeClassifier(criterion="entropy")
model = model.fit(X_train, y_train)

# Predição e Resultados
result = model.predict(X_test)
acc = metrics.accuracy_score(result, y_test)
show = round(acc * 100)
print("{}%".format(show))

print(list(result))
print(list(y_test))
```

```python
tree.plot_tree(model,filled=True)
```

```python
#segundo dataset
url_2 = "/content/sample_data/Dry_Bean_Dataset.csv"
# Carregar base de dados
# DataFrame
df_2 = pd.read_csv(url_2)

#separação dos dados em target(y) e os outros x
y_2 = df_2["Class"]
X_2 = df_2.drop("Class", axis=1)

#separando os dados de teste e treinamento
X2_train, X2_test, y2_train, y2_test = train_test_split(X_2, y_2, test_size=0.2, random_state=None, stratify=y_2) # 80% treino e 20% teste

# Treinamendo da Árvore de Decisão
model_2 = tree.DecisionTreeClassifier(criterion="entropy")
model_2 = model_2.fit(X2_train, y2_train)

# Predição e Resultados
result_2 = model_2.predict(X2_test)
acc_2 = metrics.accuracy_score(result_2, y2_test)
show_2 = round(acc_2 * 100)
print("{}%".format(show_2))

print(list(result_2))
print(list(y2_test))
```
